```python
import moviepy.editor as mp
import os
```

```python
def mp4_to_mp3(video_path, audio_path):
    """
    Extracts audio (.mp3) from video (.mp4)
    Args:
        video_path (str): the path of the video file
        audio_path (str): the path of the audio file that extracted from the video file
    """
    print(f"Extracting audio from {video_path}...")
    try:
        clip = mp.VideoFileClip(video_path)
        clip.audio.write_audiofile(audio_path)
        clip.close()
        print(f"Audio saved to {audio_path}")
    except Exception as e:
        print(f"Error during audio extraction: {e}")
        return
```

```python
mp4_dir = "data/original/phishing/mp4"
mp4_files = [f for f in os.listdir(mp4_dir) if f.endswith(".mp4")]
```

```python
mp3_dir = "data/original/phishing"
for i in range(len(mp4_files)):
    file = mp4_files[i]
    mp4_to_mp3(os.path.join(mp4_dir, file), os.path.join(mp3_dir, f"data_mp4_{i+1}.mp3"))
```
